```python
people = {
    "first":["corey", "Jane", "John"],
    "last":["Schaffer", "Jane", "Doe"],
    "email":["corey@gmail.com", "JaneDoe@gmail.com", "John@gmail.com"],
}

pp = pd.DataFrame(people)

pp
```

```python
# apply
# map
# applymap
# replace
```

```python
# apply - used to call a function on a value.Can work on either a dataframe or a series object
#  1. On series objects. When we run a function using apply, it goes to the values of the series
pp['email'].apply(len)
```

```python
def update(email):
    return email.upper()

pp['email'].apply(update)
```

```python
pp['email'] = pp['email'].apply(update)
pp
```

```python
# Use lambda
pp['email'] = pp['email'].apply(lambda x: x.lower())
```

```python
pp
```

```python
# 2. Apply on dataframe objects. When we run apply on a dataframe, it runs a function on each row or column of that df

pp['email'].apply(len)
```

```python
# will give us length of every column in a dataframe
# default axis here is rows
pp.apply(len)
```

```python
pp.apply(len, axis="columns")
```

```python
# similar to above
len(pp['email'])
```

```python
# lambda works on a series object not value
pp.apply(lambda x: x.min())
```

```python
# running apply on a series applies a function to every value in the series
# running apply to a dataframe applies a function to every series in the dataframe
# apply map only works on a datframe and applies a function to every individual element in the dataframe
```

```python
pp.applymap(len)
```

```python
pp
```

```python
pp.applymap(str.upper)
```

```python
# map method only works on a series and it works by substituting each value in a series with another value

pp['first'].map({'COREY':'CHRIS', 'JANE':'Cynthia'})
```
